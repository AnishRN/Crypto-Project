```python
import plotly.express as px

# Download historical stock data
ticker = input('Enter Ticker: ')
df = yf.download(ticker, start=start, end=end)
df = df.reset_index()

# Create an interactive line plot with hover data using Plotly Express
fig = px.line(df, x='Date', y=['Open', 'Close'], labels={'Date': 'Date', 'value': 'Price'}, title='Stock Prices')

# Show the plot
fig.show()
```

```python
# Assuming you have already downloaded and prepared your DataFrame 'df'

# Rename columns to 'ds' and 'y' for Prophet
df.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)

# Initialize Prophet model
mod = Prophet()

# Fit the model to your data
mod.fit(df)

# Create a DataFrame for future predictions
future = mod.make_future_dataframe(periods=30)

# Make predictions
forecast = mod.predict(future)

# Display the first few rows of the forecast
forecast.tail()
```

```python
# Assuming you have already obtained the 'forecast' DataFrame from the Prophet model

# Create an interactive line plot using Plotly Express
fig = px.line(forecast, x='ds', y='yhat', labels={'ds': 'Date', 'yhat': 'Forecasted Price'}, title='Stock Price Forecast')

# Show the plot
fig.show()
```

```python
fig = px.bar(df, x='ds', y=['Volume'], labels={'Date': 'Date', 'value': 'Volume'}, title='Stock Volumes')

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Candlestick(
                x=df['ds'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['y'])])

# Customize the candlestick chart layout
fig.update_layout(
    title='Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price',
    xaxis_rangeslider_visible=True
)

# Show the chart
fig.show()
```
